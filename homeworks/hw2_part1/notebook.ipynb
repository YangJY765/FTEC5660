```python
# ==============================================================================
# Smart Company Matching
# ==============================================================================
import json
import re
import nest_asyncio
from google.colab import userdata
from langchain_core.messages import HumanMessage, SystemMessage, ToolMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_mcp_adapters.client import MultiServerMCPClient

nest_asyncio.apply()

# --- PART A: SETUP ---
if 'llm' not in globals():
    GEMINI_VERTEX_API_KEY = userdata.get('VERTEX_API_KEY')
    llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GEMINI_VERTEX_API_KEY, temperature=0)
    client = MultiServerMCPClient({
        "social_graph": {
            "transport": "http",
            "url": "https://ftec5660.ngrok.app/mcp",
            "headers": {"ngrok-skip-browser-warning": "true"}
        }
    })

if 'mcp_tools' in globals():
    llm_with_tools = llm.bind_tools(mcp_tools)

# --- PART B: THE "SMART JUDGE" PROMPT ---
system_prompt = """
You are a KYC Verification Agent. Your goal is to find the person and Verify their claims.

### PHASE 1: SEARCH STRATEGY
1. **Aggressive Search**:
   - Start with `search_linkedin_people(q=NAME, location=CITY)`.
   - If 0 results, **REMOVE THE LOCATION** and search by Name only.
   - Use `fuzzy=True` for typos.
   - Try Facebook if LinkedIn fails.

### PHASE 2: VERIFICATION & SCORING (Crucial)
Once you find the profile, compare the **Company Names** in the CV vs Profile.

**RULES FOR COMPANY MATCHING:**
- **Be Smart**: Allow for abbreviations and partial matches.
  - "PwC" == "PricewaterhouseCoopers" (MATCH)
  - "Tencent" == "Tencent Holdings Ltd" (MATCH)
  - "Google" == "Google Cloud" (MATCH)
- **Do NOT reject** if the company name is just written slightly differently.

**SCORING RUBRIC:**
- **1.0 (Pass)**:
   - Profile found.
   - The recent companies in the CV **match** the Profile (semantically).
   - Minor title differences are okay (e.g., "Senior Dev" vs "Dev").

- **0.5 (Suspicious)**:
   - Profile found and Company matches.
   - BUT major discrepancies in **Job Title** or **Time**.
     * Example: CV claims "VP" (Vice President), Profile says "Intern" or "Associate".
     * Example: CV claims "PhD", Profile lists no university degree.

- **0.0 (Fail/Fake)**:
   - Scenario A: No profile found after all attempts.
   - Scenario B: **TOTAL FABRICATION**. The Profile exists, but the **Key Companies** listed on the CV are **NOT** present in the Profile at all.
     * Example: CV says "Senior Manager at Microsoft", but the Profile only shows experience at "Starbucks" or "University Student".
     * Logic: If the companies are completely different industries/entities, score 0.0.

### OUTPUT
JSON ONLY: {"candidate_name": "...", "score": 0.0, "reasoning": "Explain why companies match or don't match..."}
"""

final_scores = []

# --- PART C: EXECUTION ---
if 'all_cvs' not in globals():
    print("❌ Error: 'all_cvs' missing.")
else:
    for index, cv_data in enumerate(all_cvs):
        filename = cv_data['file']
        cv_text = cv_data['text']
        print(f"\n[{index+1}/{len(all_cvs)}] Processing {filename}...")

        messages = [
            SystemMessage(content=system_prompt),
            HumanMessage(content=f"Verify this CV. Be smart about company abbreviations.\n\n{cv_text}")
        ]

        final_score = 0.0

        for step in range(12):
            try:
                response = await llm_with_tools.ainvoke(messages)
                messages.append(response)

                if response.tool_calls:
                    print(f"  --> Step {step}: Tools: {[t['name'] for t in response.tool_calls]}")
                    for tool_call in response.tool_calls:
                        selected_tool = next((t for t in mcp_tools if t.name == tool_call["name"]), None)
                        if selected_tool:
                            # Argument Cleaning
                            tool_args = tool_call["args"]
                            for key in ["person_id", "user_id", "id"]:
                                if key in tool_args and isinstance(tool_args[key], str):
                                    numbers = re.findall(r'\d+', tool_args[key])
                                    if numbers: tool_args[key] = int(numbers[0])

                            try:
                                tool_result = await selected_tool.ainvoke(tool_args)
                                messages.append(ToolMessage(
                                    content=json.dumps(tool_result),
                                    name=tool_call["name"],
                                    tool_call_id=tool_call["id"]
                                ))
                            except Exception as e:
                                messages.append(ToolMessage(
                                    content=f"Error: {str(e)}",
                                    name=tool_call["name"],
                                    tool_call_id=tool_call["id"]
                                ))
                else:
                    final_answer = response.content
                    try:
                        json_match = re.search(r"\{.*\}", final_answer.replace("\n", ""), re.DOTALL)
                        if json_match:
                            data = json.loads(json_match.group(0))
                            final_score = float(data.get("score", 0.0))
                            reason = data.get("reasoning", "None")
                            print(f"  ✅ Decision: {reason} \n  ⭐️ Score: {final_score}")
                        else:
                            print("    ⚠️ No JSON. Score: 0.0")
                    except:
                        pass
                    break
            except Exception as e:
                print(f"Critical Error: {e}")
                break

        final_scores.append(final_score)

    scores = final_scores
    print(f"\nFinal Scores List: {scores}")
```

# Evaluation code

In the test phase, you will be given 5 CV files with fixed names:

    CV_1.pdf, CV_2.pdf, CV_3.pdf, CV_4.pdf, CV_5.pdf

Your system must process these CVs and output a list of 5 scores,
one score per CV, in the same order:

    scores = [s1, s2, s3, s4, s5]

Each score must be a float in the range [0, 1], representing the
reliability or confidence that the CV is valid (or meets the task criteria).

The ground-truth labels are binary:

    groundtruth = [0 or 1, ..., 0 or 1]

Each CV is evaluated independently using a threshold of 0.5:

- If score > 0.5 and groundtruth == 1 → Full credit
- If score ≤ 0.5 and groundtruth == 0 → Full credit
- Otherwise → No credit

In other words, 0.5 is the decision threshold.

- Each CV contributes equally.
- Final score = (number of correct decisions) / 5


```python
# =====================================================
#  Evaluation code
# =====================================================

def evaluate(scores, groundtruth, threshold=0.5):
    """
    scores: list of floats in [0, 1], length = 5
    groundtruth: list of ints (0 or 1), length = 5
    """
    assert len(scores) == 5
    assert len(groundtruth) == 5

    correct = 0
    decisions = []

    for s, gt in zip(scores, groundtruth):
        pred = 1 if s > threshold else 0
        decisions.append(pred)
        if pred == gt:
            correct += 1

    final_score = correct / len(scores)

    return {
        "decisions": decisions,
        "correct": correct,
        "total": len(scores),
        "final_score": final_score
    }
```
